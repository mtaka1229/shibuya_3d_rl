```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータの作成
np.random.seed(0)
date_range = pd.date_range(start='2023-07-01', periods=100, freq='H')
congestion_data = np.random.randint(0, 100, size=(100,))

data = pd.DataFrame({
    'Timestamp': date_range,
    'Congestion': congestion_data
})

```

```python
# 可視化
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['Congestion'], label='Congestion Level')
plt.xlabel('Time')
plt.ylabel('Congestion Level')
plt.title('Time Series Congestion Level in the Building')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Seabornを使った可視化
plt.figure(figsize=(12, 6))
sns.lineplot(x='Timestamp', y='Congestion', data=data)
plt.xlabel('Time')
plt.ylabel('Congestion Level')
plt.title('Time Series Congestion Level in the Building')
plt.xticks(rotation=45)
plt.show()
```

```python
# 複数のベクトルの定義
origins = np.array([[0, 0, 0], [1, 1, 1], [2, 2, 2]])
vectors = np.array([[1, 1, 2], [-1, -1, -2], [0, 0, 1]])

# Figureと3Dプロットの準備
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ベクトルのプロット
for origin, vector in zip(origins, vectors):
    ax.quiver(origin[0], origin[1], origin[2], 
              vector[0], vector[1], vector[2], 
              color=np.random.rand(3,), arrow_length_ratio=0.1)

# 軸ラベル
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# プロットの範囲
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

plt.show()
```
