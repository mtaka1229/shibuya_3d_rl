```python
import pandas as pd
import numpy as np

# サンプルデータの作成（実際のデータに合わせて変更してください）
data = {'t': [0, 1, 2, 3, 0],
        'k': [1, 2, 1, 3, 2]}

df_res_all = pd.DataFrame(data)

# サンプルデータをNumPy配列に変換
res_all = df_res_all.to_numpy()

# インデックス操作のベクトル演算に変更
timesteps = res_all[:, 0].astype(int)
links = res_all[:, 1].astype(int)

vizu_all = np.zeros((max(timesteps) + 1, max(links) + 1), dtype=int) # 普通にvizu_all = np.zeros((T, L))と同じ
T = 4
L = 3
# vizu_all = np.zeros((T, L))
np.add.at(vizu_all, (timesteps, links), 1)

# vizu_allの結果を表示
print(vizu_all)
```

```python
vizu_allw = np.zeros((T, L))
for i in range(len(df_res_all)):
    t = int(df_res_all.loc[i, 't']) # df_res_allでのtが0~10で振っているのでtは0~10．なのでそのままindexとして使える
    k = int(df_res_all.loc[i, 'k']) # kはid，indexはk-1．
    vizu_allw[t, k-1] += 1 ## 1列目がtimestepなのでindexとlink id は同じ

print(vizu_allw)
```
