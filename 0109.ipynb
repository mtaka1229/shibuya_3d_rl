```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 仮のCEm関数の定義（実際のCEm関数が未定義なので仮の関数としています）
def CEm(x):
    return x[0]**2 + x[1]*10   # 仮の関数例

b_k = np.array((2, 5))  ## betaは固定しておく！

x_values = np.linspace(0, 1, 10)
beta_values = np.linspace(-5, 0, 10)

CEm_values = []

for xval in x_values:
    for beta_val in beta_values:
        x = np.array([xval, beta_val])
        CEm_val = CEm(x)
        CEm_values.append(CEm_val)

# x_values, beta_values, CEm_valuesを3Dプロットする
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x_values, beta_valuesをメッシュグリッドに変換する
X, Y = np.meshgrid(x_values, beta_values)
Z = np.array(CEm_values).reshape(X.shape)

# 3Dプロットを作成する
ax.plot_surface(X, Y, Z, cmap='viridis')

# 軸ラベルの設定
ax.set_xlabel('xval')
ax.set_ylabel('beta_val')
ax.set_zlabel('CEm_values')

# グラフの表示
plt.show()
```
