```python
import pandas as pd 
import numpy as np 
import os 

date_stamp = '20230130_14'

df_node_multi = pd.read_csv('/Users/takahiromatsunaga/res2023/shibuya_nw/shibuya_stanw/ver0707/multi_node_post.csv')
nodeid_list_multi = sorted(df_node_multi['nodeid'].unique())
nodeid_list_array_multi = np.array(nodeid_list_multi)
multi_tansit_probs = pd.read_csv('/Users/takahiromatsunaga/res2023/PPcameraTG/PP/PP_webq/q_ans/multi_transit_probability.csv', index_col=0)
```

```python
# あれば
# ファイルパスを定義
file_path = f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/transit_probability_{date_stamp}30_20sec_off_ver0826.csv'
gate_nodeids = [28, 29, 35, 36, 39]
nodeid_list = sorted(df_node_multi['nodeid'].unique())
gate_nodeixs = [nodeid_list.index(item) for item in gate_nodeids]
print(gate_nodeixs)
# ファイルが存在するかチェック
if os.path.exists(file_path):
    transit_cond_prob = pd.read_csv(file_path, index_col=0)
    fusioned_cond_probs = pd.DataFrame(0, index=nodeid_list_array_multi, columns=nodeid_list_array_multi, dtype=int)

    print(multi_tansit_probs.index, multi_tansit_probs.columns)
    print(transit_cond_prob.index, transit_cond_prob.columns)

    multi_tansit_probs.update(transit_cond_prob) #, index_col=0)

    #print(multi_tansit_probs)
    multi_tansit_probs.to_csv(f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/fusioned_condprobs_{date_stamp}30_20sec_off.csv')

else:
    # なければtransit_countファイルからやる
    transition_counts = pd.read_csv(f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/transit_counts_{date_stamp}30_20sec_off_ver0826.csv', index_col=0)
    transition_counts.iloc[gate_nodeixs, :] = 0
    transition_counts.to_csv(f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/transit_counts_{date_stamp}30_20sec_off_ver0826__.csv')

    # 条件付き遷移確率
    transit_prob = transition_counts.div(transition_counts.sum(axis=1), axis=0).fillna(0)
    transit_prob.to_csv(f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/transit_probability_{date_stamp}30_20sec_off_ver0826.csv')

    print('dsffdd', transit_prob.iloc[gate_nodeixs, :])
    print('28', transit_prob.iloc[19, :])


    print(multi_tansit_probs.index, multi_tansit_probs.columns)
    print(transit_prob.index, transit_prob.columns)

    multi_tansit_probs.update(transit_prob) #, index_col=0)

    #print(multi_tansit_probs)
    multi_tansit_probs.to_csv(f'/Users/takahiromatsunaga/res2023/bledata/ble_timegroup/{date_stamp}/fusioned_condprobs_{date_stamp}30_20sec_off_ver0826.csv')
```
