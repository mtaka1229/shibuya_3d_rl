```python
# dataframe作るなら辞書res_dictを作って
# res_dict = {}
nodeid_list = sorted(df_node['nodeid'].unique())# print(destnode_list)

gatenode_list = [28, 29, 35, 36, 39]
destnode_list = [i for i in range(51, 61)]

# gxlist = [nodeid_list.index(item) for item in gatenode_list]
# dxlist = [nodeid_list.index(item) for item in destnode_list]

# print(nodeid_list)
# def utility_gate(dist, level_dum, point, hachi_dum, minami_dum):
#     inst = dist/1000 * (-5.29) + level_dum * 0.26 + point/10 * 0.38 + hachi_dum * 1.50 + minami_dum * 1.31
#     return inst

## dataframeじゃなくてnumpyでそのまま本コードに組み込んだ方がいいのではという感じ．
def calc_gatevalue(d_ix, g_ix):
    d_loc = np.array([df_node.loc[d_ix, 'x'], df_node.loc[d_ix, 'y'], df_node.loc[d_ix, 'floor']])
    g_loc = np.array([df_node.loc[g_ix, 'x'], df_node.loc[g_ix, 'y'], df_node.loc[g_ix, 'floor']])
    g_id = nodeid_list[g_ix]
    dist = np.linalg.norm(d_loc - g_loc)
    d_floor = d_loc[2]
    g_floor = g_loc[2]
    level_dum = 0
    hachi_dum = 0
    minami_dum = 0
    print(g_id)
    if g_id == 28 or g_id == 29:
        hachi_dum = 1
    elif g_id == 35 or g_id == 36:
        minami_dum = 1
    # hachi_dum = np.isin(g_ids, [28, 29]).astype(int)
    # minami_dum = np.isin(g_ids, [35, 36]).astype(int)
    
    if d_floor == g_floor:
        level_dum = 1
    
    point = 0 # とりあえず0にしておく
    
    # gateMNL由来の効用関数に入れる
    # inst = dist/1000 * (-5.29) + level_dum * 0.26 + point/10 * 0.38 + hachi_dum * 1.50 + minami_dum * 1.31
    inst = (dist / 1000 * -5.29 +
            level_dum * 0.26 +
            point / 10 * 0.38 +
            hachi_dum * 1.2 
            # minami_dum * 1.31
            ) 
    
    return inst
 
N = len(df_node) # 吸収を含んだ全ノードの数
gateval_mat = np.zeros((N, N))

for g_id in gatenode_list:
    g_ix = nodeid_list.index(g_id)
    g_loc = (df_node.loc[g_ix, 'x'], df_node.loc[g_ix, 'y'], df_node.loc[g_ix, 'floor'])
    for d_id in destnode_list:
        d_ix = nodeid_list.index(d_id)
        d_loc = (df_node.loc[d_ix, 'x'], df_node.loc[d_ix, 'y'], df_node.loc[d_ix, 'floor'])
        gd_inst = np.exp(calc_gatevalue(d_ix, g_ix))
        gateval_mat[g_ix, d_ix] = gd_inst
# df_res = pd.DataFrame(res_dict) # とするのがスマート
```
